```python
def f(x):
    return (x+3)*(x+1)*(x-2)
x = np.linspace(-4,4,100)
s= 2
plt.plot(x,f(x))
plt.figsize=(1,1)
plt.axis((-2,4,-20,20))
plt.xlabel('$S=s^2/T$')
plt.legend(['$q^2-4\sigma^2$'])
plt.axvspan(s, 14, alpha=0.3, color='green')
plt.axvspan(-20, s, alpha=0.3, color='red')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```
