```python
lbda = Symbol('\lambda')
f = Symbol('f_0')
F = Symbol('F_0')
s = Symbol('S')
Y = Symbol('Y')
```

```python
#DISCRIMINANT DE Q
```

```python
a = (lbda+f)
b = 2*f*(f+lbda-F)
c = -3*F*f*(lbda + 2*f)
d = 4*(f*F)**2
Q = a*Y**3 + b*Y**2 + c*Y + d
Q = Poly(Q,gens=Y)
Q
```

```python
discriminant(Q,Y)
```

```python
Delta_Q = b**2*c**2 + 18*a*b*c*d -27*a**2*d**2 -4*a*c**3 -4*b**3*d
Delta_Q.expand()
```

```python
#DISCRIMINANT DE P_3
```

```python
a = (lbda+f)
b = f*(f+lbda-F)
c = -F*f*(lbda + 2*f)
d = (f*F)**2
P = a*Y**3 + b*Y**2 + c*Y + d
P = Poly(P,gens=Y)
```

```python
Delta_P = discriminant(P)
Delta_P.expand()
```

```python
a = 1
b = s*(lbda+f)
c = s*f*(f+lbda-F)
d = -s*F*f*(lbda + 2*f)
e = s*(f*F)**2
A = a*Y**4 + b *Y**3 + c * Y **2 + d *Y + e 
A = Poly(A,gens=Y)
```
